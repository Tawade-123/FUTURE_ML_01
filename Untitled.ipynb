```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from prophet import Prophet
from sklearn.metrics import mean_absolute_error
```

```python
# Load datasets
train = pd.read_csv('train.csv', parse_dates=['date'])
stores = pd.read_csv('stores.csv')
transactions = pd.read_csv('transactions.csv', parse_dates=['date'])
oil = pd.read_csv('oil.csv', parse_dates=['date'])
```

```python
# Merge datasets
train = train.merge(stores, on='store_nbr', how='left')
train = train.merge(transactions, on=['date', 'store_nbr'], how='left')
train = train.merge(oil, on='date', how='left')

train.head()
```

```python
# Daily total sales
daily_sales = train.groupby('date')['sales'].sum().reset_index()
```

```python
# Plot sales trend
plt.figure(figsize=(15, 5))
sns.lineplot(data=daily_sales, x='date', y='sales')
plt.title('Total Daily Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
# Rename columns for Prophet
sales_data = daily_sales.rename(columns={'date': 'ds', 'sales': 'y'})
```

```python
# Check structure
sales_data.head()
```

```python
# Initialize and train the model
model = Prophet()
model.fit(sales_data)
```

```python
# Create future dates
future = model.make_future_dataframe(periods=90)
```

```python
# Predict future sales
forecast = model.predict(future)
```

```python
import os

# Create the directory if it doesn't exist
os.makedirs('../output', exist_ok=True)

# Save forecast to CSV
forecast.to_csv('../output/forecast_output.csv', index=False)
```

```python
# Show forecast data
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
# Forecast plot
fig1 = model.plot(forecast)
plt.title('Sales Forecast for Next 90 Days')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
# Trend + seasonality
fig2 = model.plot_components(forecast)
plt.show()
```

```python
# Merge actuals for evaluation
merged = sales_data.merge(forecast[['ds', 'yhat']], on='ds', how='left')
```

```python
# Calculate MAE (on training data)
mae = mean_absolute_error(merged['y'], merged['yhat'])
print(f"Mean Absolute Error: {mae:.2f}")
```

```python
# Export forecast results to CSV
forecast.to_csv('forecast_output.csv', index=False)
print("Exported forecast_output.csv for Power BI use.")
```
